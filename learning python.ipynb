```python
data = np.column_stack((
    np.random.randint(1, 101, 100),
    np.random.rand(100),
    np.random.randint(50, 151, 100)
))

df = pd.DataFrame(data, columns=['Integer_1_100', 'Floats_0_1','Integers_50_150'])
df.to_csv('samples_data.csv', index=False)
```

```python
sd = pd.read_csv('samples_data.csv')
print(sd.head())
print(sd.describe())
print(sd.dropna())
```

```python
import matplotlib.pyplot as plt

plt.plot(sd['Integer_1_100'], label='Random ints')
plt.plot(sd['Floats_0_1'], label='Random float')
plt.title('Samples Data plot')
plt.xlabel('Index')
plt.ylabel('values')
plt.legend()
plt.grid()
plt.show
```
